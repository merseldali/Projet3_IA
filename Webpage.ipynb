```python
import numpy as np
import pandas as pd

import ipywidgets as widgets

from IPython.display import clear_output, display, HTML
from IPython.display import set_matplotlib_formats

import warnings
warnings.filterwarnings(action='ignore')
```

```python
#df = pd.read_csv("/content/data_cleaned.csv", sep=";")

url='https://drive.google.com/file/d/1vlPuTzNWjajSAD5WoW46_ngXM2NAPuQh/view?usp=sharing'
url='https://drive.google.com/uc?id=' + url.split('/')[-2]
df = pd.read_csv(url, sep=";")

df['code_str'] = df['code'].map(str)
```

```python
class OpenFoodFactsApp:
  def __init__(self, df) -> None:
      self._df = df

      self._data = np.nan

      self._wt1 = widgets.HTML(value="<h1><center style=\"color:blue\"> Global Score App </center></h1> <center><img src=\"openfoodfacts-logo.png\"/></center> <center> <b> Scannez votre produit, analysez, comparez et achetez consciemment !</b></center>")

      self._code_label = widgets.Label(value="Code-barre du produit :")
      self._data_found_label = widgets.Label(value="")

      self._wpt1 = "Nom du produit : "
      self._wpt2 = "Catégorisation du produit : "
      self._wpt3 = "Nutriscore : "
      self._wpt4 = "Ecoscore : "
      self._wpt5 = "G-score : "
      self._wpt6 = "Classement du produit dans sa catégorie : "
      self._wpt7 = "Meilleur produit dans sa catégorie : "

      self._wp1 = widgets.HTML(value = f"<b><font color='green'>{self._wpt1}</b>")
      self._wp2 = widgets.Label(value="")
      self._wp3 = widgets.HTML(value = f"<b><font color='green'>{self._wpt2}</b>")
      self._wp4 = widgets.Label(value="")
      self._wp5 = widgets.HTML(value = f"<b><font color='green'>{self._wpt3}</b>")
      self._wp6 = widgets.Label(value="")
      self._wp7 = widgets.HTML(value = f"<b><font color='green'>{self._wpt4}</b>")
      self._wp8 = widgets.Label(value="")
      self._wp9 = widgets.HTML(value = f"<b><font color='green'>{self._wpt5}</b>")
      self._wp10 = widgets.Label(value="")
      self._wp11 = widgets.HTML(value = f"<b><font color='green'>{self._wpt6}</b>")
      self._wp12 = widgets.Label(value="")
      self._wp13 = widgets.HTML(value = f"<b><font color='green'>{self._wpt7}</b>")
      self._wp14 = widgets.Label(value="")


      self._wd1 = widgets.HTML(value = f"<hr/>")
      self._wd2 = widgets.HTML(value = f"<hr/>")

      self._box_layout = widgets.Layout(display='flex',flex_flow='column', align_items='center', width='100%')

      self._w_search = self._createSearchButton()
      self._w_product_input = self._createProductInput()

      self._title_container = widgets.VBox([self._wt1])

      self._input_container = widgets.VBox(children=[widgets.HBox([self._code_label, self._w_product_input]), widgets.HBox([self._w_search, self._data_found_label])], layout=self._box_layout)

      self._product_container = widgets.VBox([widgets.HBox([self._wp1, self._wp2, self._wp3, self._wp4 ]), widgets.HBox([self._wp5, self._wp6, self._wp7, self._wp8, self._wp9, self._wp10  ]), widgets.HBox([self._wp11, self._wp12]), widgets.HBox([self._wp13, self._wp14])])

      self.main_container = widgets.VBox(children=[self._title_container,self._wd1, self._input_container, self._wd2, self._product_container], layout=self._box_layout)

  @classmethod
  def create_class(cls, df):
      return cls(df)

  def _createProductInput(self):
    widget_code = widgets.Textarea(
      value='',
      placeholder='Entrez le code-barre',
      description='',
      disabled=False
    )
    return widget_code

  def _createSearchButton(self):
    widget_search = widgets.Button(
      description='Rechercher',
      disabled=False,
      button_style='', # 'success', 'info', 'warning', 'danger' or ''
      tooltip='Rechercher',
      icon='check' # (FontAwesome names without the `fa-` prefix)
    )
    widget_search.on_click(self._callback_search)
    return widget_search

  def _callback_search(self, btn):
    if (self._df.loc[df['code_str'] == str(self._w_product_input.value),:].shape[0] > 0):
      self._data = self._df.loc[df['code_str'] == str(self._w_product_input.value),:]
      self._data_found_label.value = 'Produit trouvé'
      self._update_app()
    else:
      self._data_found_label.value = 'Produit introuvable'
      self._clear_app()
  
  def _update_app(self):
    self._wp2.value = self._data.iloc[0]['product_name']
    self._wp4.value = str(self._data.iloc[0]['pnns_groups_1']) + ' / ' + str(self._data.iloc[0]['pnns_groups_2'])
    self._wp6.value = str(self._data.iloc[0]['nutrition_grade_fr']).upper()
    self._wp8.value = str(self._data.iloc[0]['eco_score_grade']).upper()
    self._wp10.value = str(self._data.iloc[0]['g_score_grade']).upper()
    self._wp12.value = str(int(self._data.iloc[0]['rank'])) + ' / ' + str(self._df.loc[self._df['pnns_groups_2'] == self._data.iloc[0]['pnns_groups_2'],:].shape[0])
    self._wp14.value = str(self._df.loc[(self._df['pnns_groups_2'] == self._data.iloc[0]['pnns_groups_2']) & (self._df['rank'] == 1),'code_str'].iloc[0])
  
  def _clear_app(self):
    self._wp2.value = ''
    self._wp4.value = ''
    self._wp6.value = ''
    self._wp8.value = ''
    self._wp10.value = ''
    self._wp12.value = ''
    self._wp14.value = ''
```

```python
app = OpenFoodFactsApp.create_class(df)

app.main_container
```
